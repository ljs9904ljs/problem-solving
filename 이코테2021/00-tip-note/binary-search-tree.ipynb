```python

class Tree:
    
    def __init__(self, val=None):
        if val is not None:
            self.val = val
            self.left = Tree()
            self.right = Tree()
        else:
            self.val = None
            self.left = None
            self.right = None
    
    
    def is_nil(self):
        return self.val is None
    
    
    def is_leaf(self):
        return (not self.is_nil()) and self.left.is_nil() and self.right.is_nil()
    
    
    def find_max_val(self):
        if self.is_leaf():
            return self.val
        else:
            return self.right.find_max_val()
    
            
    def insert(self, val):
        if self.is_nil():
            self.val = val
            self.left = Tree()
            self.right = Tree()
            return True
        
        if val < self.val:
            return self.left.insert(val)
        else:
            # 중복된 값은 오른쪽에 저장한다.
            return self.right.insert(val)

    
    def search(self, val):
        if self.is_nil():
            return False
        
        if val == self.val:
            print("find!")
            return True
        elif val < self.val:
            return self.left.search(val)
        else:
            return self.right.search(val)
    
    
    def delete(self, val):
        if self.is_nil():
            return False
        
        if val < self.val:
            return self.left.delete(val)
        elif val > self.val:
            return self.right.delete(val)
        else:
            # case 1) 제거하려는 node의 left child와 right child가 둘 다 살아 있을 때
                # max(left child subtree) 혹은 min(right child subtree)인 노드를 제거하려는 node에 가져다 넣어야 함.
            # case 2) left child나 right child 중에 하나만 살아 있을 때
            # case 3) child가 없을 때
            
            if self.is_leaf():
                self.val = None
                self.left = None
                self.right = None
            elif self.left.is_nil():
                self.val = self.right.val
                self.left = self.right.left
                self.right = self.right.right
            else:
                # max_val이 중복된 값으로 여러 개 들어 있더라도 괜찮다. 잘 처리된다.
                max_val = self.left.find_max_val()
                self.val = max_val
                
                # max(left subtree)를 찾아서 그 노드를 제거하고 값은 self node로 가져온다.
                    # 중복된 값이 있다면 최대한 leaf 쪽으로 이동한다.
                cur = self.left
                while not cur.is_nil():
                    if cur.val == max_val:
                        while (not cur.is_leaf()) and cur.val == cur.right.val:
                            cur = cur.right
                        cur.delete(max_val)
                        break
                    cur = cur.right
                
            return True
        
    
    def traverse(self):
        # inorder traversal (left -> this -> right)
        
        if self.is_nil():
            return []
        else:
            return self.left.traverse() + [self.val] + self.right.traverse()

        
            
```
