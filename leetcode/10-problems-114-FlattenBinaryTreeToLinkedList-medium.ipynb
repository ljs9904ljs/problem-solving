```python
from typing import Optional

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
    
    
    def __str__(self):
        return f"TreeNode({self.val})"
        
        
    def preorder_traversal(self):
        if self.val == 0: return []
        
        result = [self.val]
        if self.left is not None:
            result += self.left.preorder_traversal()
        
        if self.right is not None:
            result += self.right.preorder_traversal()
        
        return result


# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def flatten(self, root: Optional[TreeNode]) -> None:
        """
        Do not return anything, modify root in-place instead.
        """
        if root is None: return None

        def is_before_leaf(t):
            if t is None: return False
            
            left_check = t.left is not None and t.left.left is None and t.left.right is None
            right_check = t.right is not None and t.right.left is None and t.right.right is None
            if t.left is not None and t.right is not None:
                return left_check and right_check
            if t.left is not None:
                return left_check
            if t.right is not None:
                return right_check

            return False


        def flat(t):
            
            # leaf node
            if t.left is None and t.right is None: return (t, t)

            if is_before_leaf(t):
                # print("\tif is_before_leaf : ", t)
                if t.left is not None:
                    t.left.right = t.right
                    t.right = t.left
                    t.left = None
                
                right_end = t
                while right_end.right is not None:
                    right_end = right_end.right
                    
                # print("\t\tbefore leaf 노드 정돈한 후: ", t.preorder_traversal())

                # 정돈된 linked list의 가장 위와 가장 아래
                return (t, right_end)
            else:
                # print("\telse is_before_leaf ")
                
                if t.left is None:
                    if t.right is None: raise Exception("unreachable")

                    right_result = flat(t.right)
                    t.right = right_result[0]
                    return (t, right_result[1])
                elif t.right is None:
                    if t.left is None: raise Exception("unreachable")

                    left_result = flat(t.left)
                    t.right = left_result[0]
                    t.left = None
                    return (t, left_result[1])
                else:
                    left_result = flat(t.left)
                    right_result = flat(t.right)
                    # print("\tleft_result: ", left_result[0].preorder_traversal(), left_result[0].val, left_result[1].val)
                    # print("\tright_result: ", right_result[0].preorder_traversal(), right_result[0].val, right_result[1].val)

                    t.right = left_result[0]
                    left_result[1].right = right_result[0]
                    t.left = None
                    return (t, right_result[1])

        flat(root)




# t = TreeNode(1)
# t2 = TreeNode(2)
# t3 = TreeNode(3)
# t4 = TreeNode(4)
# t5 = TreeNode(5)
# t6 = TreeNode(6)

# t.left = t2
# t.right = t5
# t2.left = t3
# t2.right = t4
# t5.right = t6

# print(t.preorder_traversal())
# s = Solution()
# s.flatten(t)
# print(t.preorder_traversal())



```

```python

"""

훨씬 깔끔한 답변을 찾았기 때문에, 그것을 아래에 추가해두었다. 

root node부터 탐색을 시작하면서, 
node의 left를 node -> node의 원래 left -> node의 원래 right 순서로 다시 연결한다.
그리고 node.right부터 위와 같은 과정을 쭉 반복한다.

"""




# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def flatten(self, root: Optional[TreeNode]) -> None:
        """
        Do not return anything, modify root in-place instead.
        """

        if not root:
            return None 

        node = root 

        while node:

            # if having a left child 
            if node.left:

                rightmost = node.left 
                while rightmost.right:
                    rightmost = rightmost.right 

                #rewire connection 
                rightmost.right = node.right 
                node.right = node.left 
                node.left = None 

            node = node.right 





```
