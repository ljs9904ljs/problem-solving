```python
from typing import Optional


# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
    
    def dump(self):
        cur = self
        while cur is not None:
            print(cur.val)
            cur = cur.next
        
        
class Solution:
    def swapPairs(self, head: Optional[ListNode]) -> Optional[ListNode]:

        # linked list의 길이가 0 또는 1.
        if head is None or head.next is None:
            return head

        # linked list의 길이가 2이상.
        prev = None
        cur = head
        head = cur.next  # 새롭게 head를 세팅해줘야 노드를 놓치지 않는다.

        while cur is not None:
            
            if cur.next is None:
                break
            else:
                temp = cur.next.next
                if prev is not None:
                    prev.next = cur.next
                cur.next.next = cur
                cur.next = temp

                if cur.next is not None:
                    prev = cur
                    cur = cur.next
                else:
                    break
        
        return head
            

n4 = ListNode(4, None)
n3 = ListNode(3, n4)
n2 = ListNode(2, n3)
n1 = ListNode(1, n2)

n1.dump()
print("===========")
            
s = Solution()
result = s.swapPairs(n1)
result.dump()


```
