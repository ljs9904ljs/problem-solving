```python
from typing import List

class Solution:
    def spiralOrder(self, matrix: List[List[int]]) -> List[int]:


        # (오른쪽, 아래, 왼쪽, 위 -> 오른쪽, 아래, 왼쪽, 위 -> ... ) 순으로 이동하면 된다.

        # 오른쪽, 아래, 왼쪽, 위
        dr = [0, 1, 0, -1]
        dc = [1, 0, -1, 0]

        # 시작 좌표
        r = 0
        c = 0

        result = []

        m = len(matrix)
        n = len(matrix[0])

        visited = [[False] * n for _ in range(m)]

        # 오른쪽? 아래? 왼쪽? 위?
        cur_movement = 0
        
        for _ in range(m * n):
            result.append(matrix[r][c])
            visited[r][c] = True

            nr = r + dr[cur_movement % 4]
            nc = c + dc[cur_movement % 4]
            if  0 <= nr < m and 0 <= nc < n and not visited[nr][nc]:
                r = nr
                c = nc
            else:
                # 방향 꺾기
                cur_movement += 1
                nr = r + dr[cur_movement % 4]
                nc = c + dc[cur_movement % 4]

                # 한 번 방향을 꺾은 직후에는 반드시 접근 가능한 martix entry이다.
                # 그래서 또 다시 범위를 체크할 필요는 없다.
                r = nr
                c = nc
        
        return result


            
        
```
