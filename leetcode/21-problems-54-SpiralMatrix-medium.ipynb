```python
from typing import List

class Solution:
    def spiralOrder(self, matrix: List[List[int]]) -> List[int]:


        # (오른쪽, 아래, 왼쪽, 위 -> 오른쪽, 아래, 왼쪽, 위 -> ... ) 순으로 이동하면 된다.

        # 오른쪽, 아래, 왼쪽, 위
        dr = [0, 1, 0, -1]
        dc = [1, 0, -1, 0]

        # 시작 좌표
        r = 0
        c = 0

        result = []

        m = len(matrix)
        n = len(matrix[0])

        visited = [[False] * n for _ in range(m)]

        # 오른쪽? 아래? 왼쪽? 위?
        cur_movement = 0
        
        for _ in range(m * n):
            result.append(matrix[r][c])
            visited[r][c] = True

            nr = r + dr[cur_movement % 4]
            nc = c + dc[cur_movement % 4]
            if  0 <= nr < m and 0 <= nc < n and not visited[nr][nc]:
                r = nr
                c = nc
            else:
                # 방향 꺾기
                cur_movement += 1
                nr = r + dr[cur_movement % 4]
                nc = c + dc[cur_movement % 4]

                # 한 번 방향을 꺾은 직후에는 반드시 접근 가능한 martix entry이다.
                # 그래서 또 다시 범위를 체크할 필요는 없다.
                r = nr
                c = nc
        
        return result


            
        
```

```python
"""
다른 사람의 풀이를 보는 도중 놀라운 solution을 찾았음...

matrix를 transpose하는 방법에 관한 아이디어도 제공해준다.

"""


from typing import List

class Solution:
    def spiralOrder(self, matrix: List[List[int]]) -> List[int]:
        
        # "matrix and" 부분은 바로 뒤의 matrix.pop(0)을 할 때 IndexError가 나는 것을 방어하며 안전하게 pop할 수 있는 방식이다.
        # list가 비어있으면 False이고 원소가 하나라도 있다면 True이다.
        return matrix and [*matrix.pop(0)] + self.spiralOrder([*zip(*matrix)][::-1])
    
            
        
```
