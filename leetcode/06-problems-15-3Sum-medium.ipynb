```python
from bisect import bisect_left, bisect_right
from typing import List

class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:

        # nums.length <=3000 이므로 O(n^3)은 안 된다. O(n^2 Log n)은 된다.
        # nums를 정렬한다.
        # 트리플렛 중 2개(a랑 b라고 하자)를 적당히 고르고, 나머지 1개(nums[a] + nums[b] + 나머지 1개 == 0, 을 만족하는 것을 찾으면 된다.)는 바이너리 서치로 찾으면 된다.
        # 중복 방지는 트리플렛을 만들 때 정렬해서 tuple로 만든 다음, set에 집어 넣으면 된다.

        def count_by_range(arr, a, b):
            # Both are inclusive.
            return bisect_right(arr, b) - bisect_left(arr, a)

        nums.sort()
        result = set()

        i = 0
        while i < len(nums):
            j = i + 1
            while j < len(nums):
                target = -(nums[i] + nums[j])
                idx = bisect_right(nums, target)
                idx -= 1
                if j < idx < len(nums) and nums[idx] == target:
                    temp = [nums[i], nums[j], nums[idx]]
                    temp.sort()
                    result.add(tuple(temp))
                
                
                j = bisect_right(nums, nums[j])  # 중복되는 값들을 한 번에 점프한다.
            i += 1  # 이 부분을 "i = bisect_right(nums, nums[i])" 로 변경하면 2482 ms, 20.8 MB가 나온다.
            

                    

        result = list(result)
        result = [list(t) for t in result]

        return result

# Leetcode Submissions 기준
    # Runtime: 2710 ms
    # Memory: 20.8 MB


# 이렇게 풀면 속도가 좀 느리기는 하다. 
# two pointers 기법을 좀 더 멋지게 활용하면 빠르게 계산할 수 있다.
        
```

```python
from bisect import bisect_left, bisect_right

class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:

        # two pointers를 잘 써보자.
        nums.sort()
        result = []

        i = 0
        while i <= (len(nums) - 1 - 1 - 1):
            j = i + 1
            k = len(nums) - 1  
            while j < k:
                v = nums[i] + nums[j] + nums[k]
                if v < 0:
                    j += 1
                elif v > 0:
                    k -= 1
                else:
                    result.append([nums[i], nums[j], nums[k] ] )
                    j = bisect_right(nums, nums[j])  # 중복되는 값은 한 번에 점프
            
            i = bisect_right(nums, nums[i])  # 중복되는 값은 한 번에 점프





        return result


# Leetcode Submissions 기준
    # Runtime: 581 ms
    # Memory: 19.7 MB


# Set 연산을 하지 않게 됨으로써 발생하는 계산적 이점이 큰 것 같다.
```
