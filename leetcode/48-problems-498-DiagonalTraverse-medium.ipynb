```python
"""
Given an m x n matrix mat, return an array of all the elements of the array in a diagonal order.

 

Example 1:


Input: mat = [[1,2,3],[4,5,6],[7,8,9]]
Output: [1,2,4,7,5,3,6,8,9]
Example 2:

Input: mat = [[1,2],[3,4]]
Output: [1,2,3,4]
 

Constraints:

m == mat.length
n == mat[i].length
1 <= m, n <= 104
1 <= m * n <= 104
-105 <= mat[i][j] <= 105

"""



```

```python
from typing import List


class Solution:
    def findDiagonalOrder(self, mat: List[List[int]]) -> List[int]:
        
        M = len(mat)
        N = len(mat[0])

        LEN = M + N - 1

        startpoints = []
        for i in range(N):
            startpoints.append((0, i) )
        
        for j in range(1, M):
            startpoints.append((j, N - 1) )
        
        def collect_diagonals_from(point):
            # point는 (row 번호, col 번호) 형식의 튜플이다.

            dr = 1
            dc = -1
            result = []  # list of Tuple[row 번호, col 번호]

            r, c = point
            result.append((r, c) )

            nr = r + dr
            nc = c + dc
            while 0 <= nr < M and 0 <= nc < N:
                result.append((nr, nc) )
                nr += dr
                nc += dc
            
            return result
        
        ret = []
        for i, p in enumerate(startpoints):
            values = [mat[r][c] for r, c in collect_diagonals_from(p)]
            if i % 2 == 1:
                ret.extend(values)
            else:
                ret.extend(values[::-1])
        
        return ret

        

```
