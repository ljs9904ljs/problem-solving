```python
""" 
Given an integer array nums, find a 
subarray
 that has the largest product, and return the product.

The test cases are generated so that the answer will fit in a 32-bit integer.

 

Example 1:

Input: nums = [2,3,-2,4]
Output: 6
Explanation: [2,3] has the largest product 6.
Example 2:

Input: nums = [-2,0,-1]
Output: 0
Explanation: The result cannot be 2, because [-2,-1] is not a subarray.
 

Constraints:

1 <= nums.length <= 2 * 104
-10 <= nums[i] <= 10
The product of any subarray of nums is guaranteed to fit in a 32-bit integer.

"""

from typing import List
import math


class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        
        """
        아래 풀이 방법에서는 for loop가 2중이지만 O(N^2 / 2)이다.
        왜냐하면, sliding window의 크기가 1일 때 N개, 2일 때 N-1개, 3일 때 N-2개, ... N일 때 1개 이기 때문에 이를 합하면 (N(N+1) / 2)번이다.
        하지만 nums.length의 최대 길이가 2 * 10^4 라서 TLE가 발생하는 건 피할 수 없는 듯하다.
        테스트 케이스 중 186개(총 190개 중에서)는 패스했다.
        
        """
        # N = len(nums)
        # max_product = -int(1e9)
        # for window_size in range(1, N + 1):
        #     for i in range(N):
        #         max_product = max(max_product, math.prod(nums[i:i + window_size]))
        
        # return max_product




        # Kadane's Algorithm을 응용한 풀이 방법이다. 종이에 적으면서 풀이를 구상했다.

        prev_max = nums[0]
        prev_min = nums[0]
        max_product = nums[0]

        for i in range(1, len(nums)):
            candidates = [prev_max * nums[i], prev_min * nums[i], nums[i]]
            prev_max = max(candidates)
            prev_min = min(candidates)
            max_product = max(max_product, prev_max)

        return max_product
```
