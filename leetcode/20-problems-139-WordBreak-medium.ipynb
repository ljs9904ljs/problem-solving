```python
from collections import deque
from typing import List
from functools import cache

class Solution:
    def wordBreak(self, s: str, wordDict: List[str]) -> bool:
        """
        wordDict 안에 들어 있는 것을 전부 다 하나씩 s의 가장 앞부분에 대보면서 일치하면 지운다.
        그리고 남은 s를 다시 대보면서 지운다. 이런 식으로 했을 때 다 지워지는 경우가 하나라도 존재한다면 true.
        아니면 false. -> 이렇게 하면 메모리 초과...
        """
        
        """
        # 아래는 메모리 초과 오류가 발생하는 방식이다...

        q = deque()
        q.append(s)
        while q:
            cur = q.popleft()
            for word in wordDict:
                if cur.startswith(word):
                    if len(cur[len(word):]) == 0:
                        return True

                    q.append(cur[len(word):])
        
        return False
        """

        # dp[i] = i번째 인덱스까지 살폈을 때 true? or false? ( 0 <= i < len(s) )
        # dp[i] = dp[i] or ( dp[i - len(word[j] ) ] and (s[i - len(word[j]) + 1: i + 1] == word[j]) )

        INF = int(1e9)  # 스트링 wordDict[i].length의 길이의 최대 값은 20이다. 적당히 큰 값을 잡았다.

        @cache
        def dp(i):
            shortest_len = INF
            for word in wordDict:
                if len(word) < shortest_len:
                    shortest_len = len(word)
                if word == s[:i+1]:
                    return True
            
            if shortest_len >= i + 1:
                return False

            # 하나라도 만족한다면 s[:i+1]은 만들 수 있는 것이다.
            for word in wordDict:
                # slicing의 특성 상 길이 체크를 한 번 더 해야 함.
                if dp(i - len(word)) and \
                    len(s[i - len(word) + 1: i + 1]) == len(word) and \
                    s[i - len(word) + 1: i + 1] == word:
                    return True

            return False

        return dp(len(s) - 1)


        
```
