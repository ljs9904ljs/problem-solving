```python
from typing import List


class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        
        # integers -> counting sort인가? counting sort가 O(n)이니까 거기에서 제일 긴 subarray를 찾으면 되는 건가..?
        # 그러기엔 값이 10억까지 될 수 있어서 너무 크다...


        # O(n log n)으로 먼저 풀어봐야 겠다.

        # DP 아이디어를 떠올리기 위한 예시들..

        # 0 1 1 5 6 7
        # 1 2 1 1 2 3

        # 0 1 1 2 5 6
        # 1 2 2 3 1 1

        # 1 2    3   4   100   200
        # 1 1+1 2+1 3+1   1     1
        
        # 0
        # 1

        # 0 0
        # 1 1

        # 정렬 후, DP로 처리하는 방식 ==>> O(n log n)

        # if len(nums) == 0:
        #     return 0

        # nums.sort()

        # table = [1] * len(nums)
        # for i in range(1, len(nums)):
        #     if nums[i - 1] + 1 == nums[i]:
        #         table[i] = table[i - 1] + 1
        #     elif nums[i - 1] == nums[i]:
        #         table[i] = table[i - 1]
        #     else:
        #         table[i] = 1
        
        # return max(table)



        # O(n) 풀이는 다른 사람의 solution을 보고 알게 되었다... 
        # 덤으로 set이 hash table로 구현되어서 insert/read/delete가 
        # 거의 O(1) (collision handling에 따라 O(n)도 됨.)이라는 것을 알게 되었다.

        # 아래 코드가 어째서 O(n)인가?
            # 모든 inner while loop의 길이를 다 더해도 O(n)이기 때문이다. ==>> 궁극적으로 O(2n) = O(n)
                # 1,2,3,4,10,11,12를 생각해보면
                # "if x-1 not in nums"를 체크하는 부분으로 인해 1과 10만 while을 수행할 가능성이 생긴다.
                # 1일 때 y == 2 -> 3 -> 4 -> 5에서 끝나고,  10일 때 y == 11 -> 12 -> 13에서 끝난다. 

                # 1 1 1 2 3 4 5 6 7 10 11 12 라면? -> set이므로 불가능하다 모든 원소는 1개씩만 구별되게 존재한다.

        nums = set(nums)  # O(n)
        result = 0
        for x in nums:  # O(n)
            if x - 1 not in nums:
                y = x + 1
                while y in nums:
                    y += 1
                result = max(result, y - x)
        
        return result  # 총 O(2n) = O(n)
```
