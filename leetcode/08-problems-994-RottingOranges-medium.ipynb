```python
from typing import List

from collections import deque
import heapq

class Solution:
    def orangesRotting(self, grid: List[List[int]]) -> int:
        
        
        # Connected component에 관한 문제이다.
        # 상하좌우로 연결되어 있어야 한다. 대각선은 연결되어 있지 않은 것이다.
        # 거리를 1씩 증가시키면서 탐색하면 되니까 BFS를 쓰면 된다.

        # 즉, BFS로 connected node들을 탐색하면서 거리 값을 계산해준 뒤, 
        # 마지막에 거리 값이 가장 큰 것을 return하면 된다.

        # 라고 생각했으나 하나의 Connected component 안에 여러 개의 썩은 오렌지가 존재할 경우 그 썩은 오렌지들에서 동시에 BFS가 퍼져나가야 한다. 따라서 이를 위해 priority queue를 사용했다.

        m = len(grid)
        n = len(grid[0])
        dist = [[0] * n for _ in range(m)]
        visited = [[False] * n for _ in range(m)]

        # 상, 하, 좌, 우 순서로
        dr = [-1, 1, 0, 0]
        dc = [0, 0, -1, 1]

        # 결과 값
        max_minutes = 0

        def bfs():
            q = []
            for i in range(m):
                for j in range(n):
                    if grid[i][j] == 2:
                        heapq.heappush(q, (dist[i][j], (i, j)))
                        visited[i][j] = True

            while q:
                priority, rc = heapq.heappop(q)
                r, c = rc

                for i in range(len(dr)):
                    nr = r + dr[i]
                    nc = c + dc[i]

                    if 0 <= nr < m and 0 <= nc < n \
                        and not visited[nr][nc] and grid[nr][nc] == 1:
                        
                        visited[nr][nc] = True
                        dist[nr][nc] = dist[r][c] + 1
                        grid[nr][nc] = 2

                        heapq.heappush(q, (dist[nr][nc], (nr, nc)))

        bfs()
        
        
        for i in range(m):
            for j in range(n):
                max_minutes = max(max_minutes, dist[i][j])
                if grid[i][j] == 1:
                    return -1
        
        return max_minutes
```
