```python
from typing import List


class Solution:
    def gameOfLife(self, board: List[List[int]]) -> None:
        """
        Do not return anything, modify board in-place instead.
        """
        
        M = len(board)
        N = len(board[0])

        """
        '@' 문양을 중심으로 1~8번까지를 순회하며 검사할 것이다.
        1 2 3
        4 @ 5
        6 7 8

        """
        dr = [-1, -1, -1, 0, 0, 1, 1, 1]
        dc = [-1, 0, 1, -1, 1, -1, 0, 1]


        def live_or_dead(r: int, c: int) -> int:
            if not (0 <= r < M and 0 <= c < N):
                # infinite한 보드 위에서 진행되는데, board 리스트의 영역 밖을 어떻게 계산할 것인가? 위키백과에서 여러 전략을 소개하고 있다.
                # 영역 밖을 dead로 간주하는 전략을 우선 채택해본다.
                return 0  

            # 아래에 '|=' 연산자를 통해 밑에서 2번째 비트에 다음 차례의 값을 저장해두는데, 그게 영향을 미치지 않도록 값을 조회할 때 '0000...0001'이랑 bit AND 연산을 해서 조회한다!
            return board[r][c] & 1


        def check(r: int, c: int) -> int:
            live_count = 0
            for i in range(len(dr)):
                live_count += live_or_dead(r + dr[i], c + dc[i])

            if live_or_dead(r, c) == 1:
                if live_count < 2:
                    return 0
                elif 2 <= live_count <= 3:
                    return 1
                else:
                    return 0
            else:
                if live_count == 3:
                    return 1
                else:
                    return 0
        

        temp = [[0] * N for _ in range(M)]
        for i in range(M):
            for j in range(N):
                # 하위 2번째 비트를 이용해서 별도의 리스트를 할당하는 것보다는 적은 메모리 사용으로 문제를 해결했다.
                board[i][j] |= (check(i, j) << 1)
        
        for i in range(M):
            for j in range(N):
                board[i][j] = (board[i][j] >> 1)


            
            

```
