```python
from typing import List

class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        """
        (처음 생각1)
        input 값의 범위를 보니 brute force로도 문제를 해결할 수 있다. n,m이 커봐야 300이기 때문에 2중 for문으로 해결해도 
        시간 초과가 발생하지 않을 것이다.


        (처음 생각2) 
        대각선을 기준으로 생각해볼 수 있다.
        Example 1처럼 생긴 matrix 안에서 6을 찾는다고 해보자.
        대각선의 값은 1, 5, 9, 17, 30이다. 이 중에서 5를 살펴보면 
        matrix의 조건으로 인해 5를 우측 하단의 코너로 하는 작은 사각형에는 5이하의 값들만 존재할 수 있다.
        그리고 9에 대해서는 마찬가지로 9를 우측 하단의 코너로 하는 작은 사각형 안에는 9이하의 값들만 존재할 수 있으므로
        6이 존재할 수도 있다. 따라서 {9에 대한 작은 사각형} - {5에 대한 작은 사각형} 인 영역에서 6을 찾아보면 된다.

        (다른 사람의 solution을 보고 나서)
        우측 상단의 entry를 root로 하는 binary search tree 비슷하게 생긴 무언가로 생각할 수 있다는 것을 알게 되었다.


        """
        # <풀이 1>
        # m = len(matrix)
        # n = len(matrix[0])
        
        # for i in range(m):
        #     for j in range(n):
        #         if matrix[i][j] == target:
        #             return True
        # return False
        

        # <풀이 2>
        m = len(matrix)
        n = len(matrix[0])

        r = 0
        c = n - 1

        while 0 <= r < m and 0 <= c < n:
            
            if matrix[r][c] == target:
                return True
            elif matrix[r][c] < target:
                r += 1
            else:
                c -= 1
        
        return False
```
