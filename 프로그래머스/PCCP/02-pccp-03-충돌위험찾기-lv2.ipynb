```python
tc1 = [ [[3, 2], [6, 4], [4, 7], [1, 4]], [[4, 2], [1, 3], [2, 4]] ]
sol1= 1

tc2 = [ [[3, 2], [6, 4], [4, 7], [1, 4]], [[4, 2], [1, 3], [4, 2], [4, 3]] ]
sol2 = 9

tc3 = [ [[2, 2], [2, 3], [2, 7], [6, 6], [5, 2]], [[2, 3, 4, 5], [1, 3, 4, 5]] ]
sol3 = 0
```

```python
def solution(points, routes):
    answer = 0
    
    class Coord:
        def __init__ (self, _num, _r, _c):
            self.num = _num
            self.r = _r
            self.c = _c
            
        def __str__ (self):
            return f"Coord( num({self.num}), r({self.r}), c({self.c}) )"
        
    def find_by_num_or_none(coords, num):
        for coord in coords:
            if coord.num == num:
                return coord
        raise Exception("발생하면 안 되는 경우임.")
    
    
    def get_path_between_houses(from_house_coord, to_house_coord):
        
        if type(from_house_coord) is not Coord or type(to_house_coord) is not Coord:
            raise Exception("[get_path_between_houses] input type error")
        
        
        diff_r = to_house_coord.r - from_house_coord.r
        diff_c = to_house_coord.c - from_house_coord.c
        
        path = []
        nr = from_house_coord.r
        nc = from_house_coord.c
        
        if diff_r > 0:
            for i in range(1, diff_r + 1):
                path.append( (nr + i, nc) )
        elif diff_r < 0:
            for i in range(1, -diff_r + 1):
                path.append( (nr - i, nc) )
        
        nr += diff_r
        
        
        if diff_c > 0:
            for i in range(1, diff_c + 1):
                path.append( (nr, nc + i) )
        elif diff_c < 0:
            for i in range(1, -diff_c + 1):
                path.append( (nr, nc - i) )
                
        nc += diff_c
        
        
        return path
        
        
    
    house_coords = []
    for i in range(len(points)):
        r, c = points[i]
        house_coords.append(Coord(i + 1, r, c))
        
    paths = []
    for route in routes:
        path = []
        for j in range(len(route) - 1):
            cur_house_num = route[j]
            next_house_num = route[j + 1]
            
            from_house_coord = find_by_num_or_none(house_coords, cur_house_num)
            to_house_coord = find_by_num_or_none(house_coords, next_house_num)
            
            if j == 0:
                path.append( (from_house_coord.r, from_house_coord.c) )
            path.extend(get_path_between_houses(from_house_coord, to_house_coord))
        
        paths.append(path)
    
    # path example -> [(3, 2), (4, 2), (4, 3), (4, 4), (4, 5), (4, 6), (4, 7)]
    # paths := [path1, path2, path3, ...]
    
    
    count = 0
    for t in range(max(map(len, paths))):
        path_t = list(map(lambda path: path[t] if 0 <= t < len(path) else (-1, -1), paths))
        # raise Exception(", ".join(map(lambda x: f"[{x[0]}, {x[1]}]", path_t)))
        
        # r, c 두 가지 값이 서로 같은 개수를 체크한다.
        candidates = list(set(path_t))
        # raise Exception(", ".join(map(lambda x: f"({x[0]}, {x[1]})", candidates)))
        d = dict()
        for candidate in candidates:
            if candidate == (-1, -1):
                continue
            
            d[candidate] = 0
            for coord in path_t:
                if coord == candidate:
                    d[candidate] += 1
            if d[candidate] >= 2:
                count += 1
    
    answer = count

    return answer

def expect(sol, answer):
    if sol != answer:
        print(f"틀렸음! 정답: {sol}, 하지만 나의 답: {answer}")
    else:
        print("정답!")


expect(sol1, solution(tc1[0], tc1[1]))
expect(sol2, solution(tc2[0], tc2[1]))
expect(sol3, solution(tc3[0], tc3[1]))
        
```

```python
from collections import defaultdict


def solution(points, routes):
    print("sol2 호출됨.")
    
    # key: (r, c, time), value: 해당 좌표를 해당 시각에 로봇이 지나간 횟수
    path = defaultdict(int)
    
    
    def point(num):
        return points[num - 1]
    
    
    def next(r, c, dst):
        dr, dc = point(dst)  # destination row, destination col
        
        
        if dr - r > 0:
            return (r + 1, c)
        elif dr - r < 0:
            return (r - 1, c)
        
        if dc - c > 0:
            return (r, c + 1)
        elif dc - c < 0:
            return (r, c - 1)
        
        raise Exception("r좌표나 c좌표 둘 다 안 바뀐다면 이상한 경우이다.")    
    
    
    for route in routes:
        time = 0
        
        r, c = point(route[0])
        # dr, dc = point(route[-1])
        
        path[(r, c, time)] += 1
        
        for i in range(len(route) - 1):
            r, c = point(route[i])
            dr, dc = point(route[i + 1])
        
            while not (r == dr and c == dc):
                time += 1
                nr, nc = next(r, c, route[i + 1])
                
                path[(nr, nc, time)] += 1
                
                r = nr
                c = nc
            
    result = 0
    for v in path.values():
        if v >= 2:
            result += 1
    
    return result
  
            
expect(sol1, solution(tc1[0], tc1[1]))
expect(sol2, solution(tc2[0], tc2[1]))
expect(sol3, solution(tc3[0], tc3[1]))
        
    
```

```python

# https://school.programmers.co.kr/learn/courses/30/lessons/340211
from collections import defaultdict

def solution(points, routes):
    print("sol3 호출됨.")
    path = defaultdict(int)

    def point(p):
        return points[p-1]

    def next(r, c, next):
        cy, cx = r, c
        ny, nx = points[next-1]

        if cy != ny:
            if cy > ny:
                return (r-1, c)
            else:
                return (r+1, c)
        else:
            if cx > nx:
                return (r, c-1)
            else:
                return (r, c+1)


    for route in routes:
        cnt = 0
        cy, cx = point(route[0])
        path[(cy, cx, cnt)] += 1
        for i in range(len(route)-1):
            n = route[i+1]
            ny, nx = point(n)
            while cy != ny or cx != nx:
                cy, cx = next(cy, cx, n)
                cnt += 1
                path[(cy, cx, cnt)] += 1

    answer = 0
    for _, v in path.items():
        if v >= 2:
            answer += 1   

    return answer


expect(sol1, solution(tc1[0], tc1[1]))
expect(sol2, solution(tc2[0], tc2[1]))
expect(sol3, solution(tc3[0], tc3[1]))
```
